```python
from autogluon.tabular import TabularDataset, TabularPredictor
import pandas as pd
```

```python
# read data into memory
train = TabularDataset('/content/drive/MyDrive/CSV-Data/train.csv')
```

```python
train.head(5)
```

```python
#select a label
label='Survived'
time_limit=60

#train

predictor=TabularPredictor(label=label).fit(train_data=train,time_limit=time_limit)
```

```python
# show all the models performace
predictor.leaderboard(silent=True)
```

```python
# test

test = TabularDataset('../content/drive/MyDrive/CSV-Data/test.csv')
test.head()
```

```python
result= pd.DataFrame(test['PassengerId'])
result[label]=predictor.predict(test)
result.tail(n=5)
```
